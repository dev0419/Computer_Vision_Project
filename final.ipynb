```python
#Creating OCR

import cv2
import numpy as np
import os
import matplotlib.pyplot as plt

# Constants and paths
LINE_WIDTH = 18

PATHS = [r'C:\Users\rsurs\OneDrive\Documents\University_Study Material\3rd year\CV_LAB\Mini_Project\data\sd\struct1',
        r'C:\Users\rsurs\OneDrive\Documents\University_Study Material\3rd year\CV_LAB\Mini_Project\data\sd\struct4',
        r'C:\Users\rsurs\OneDrive\Documents\University_Study Material\3rd year\CV_LAB\Mini_Project\data\sd\struct5',
        r'C:\Users\rsurs\OneDrive\Documents\University_Study Material\3rd year\CV_LAB\Mini_Project\data\sd\struct8',
        r'C:\Users\rsurs\OneDrive\Documents\University_Study Material\3rd year\CV_LAB\Mini_Project\data\sd\struct13',
        r'C:\Users\rsurs\OneDrive\Documents\University_Study Material\3rd year\CV_LAB\Mini_Project\data\sd\struct16',
        r'C:\Users\rsurs\OneDrive\Documents\University_Study Material\3rd year\CV_LAB\Mini_Project\data\sd\struct19',
        r'C:\Users\rsurs\OneDrive\Documents\University_Study Material\3rd year\CV_LAB\Mini_Project\data\sd\struct20',
        r'C:\Users\rsurs\OneDrive\Documents\University_Study Material\3rd year\CV_LAB\Mini_Project\data\sd\struct22']
# Updated absolute paths for templates
TEMPLATES = [
    r"C:\Users\rsurs\OneDrive\Documents\University_Study Material\3rd year\CV_LAB\Mini_Project\train\train\oh\combined.png\\",
    r"C:\Users\rsurs\OneDrive\Documents\University_Study Material\3rd year\CV_LAB\Mini_Project\train\train\or\combined.png\\",
    r"C:\Users\rsurs\OneDrive\Documents\University_Study Material\3rd year\CV_LAB\Mini_Project\train\train\ro\combined.png\\",
    r"C:\Users\rsurs\OneDrive\Documents\University_Study Material\3rd year\CV_LAB\Mini_Project\train\train\h\combined.png\\",
    r"C:\Users\rsurs\OneDrive\Documents\University_Study Material\3rd year\CV_LAB\Mini_Project\train\train\n\combined.png\\",
    r"C:\Users\rsurs\OneDrive\Documents\University_Study Material\3rd year\CV_LAB\Mini_Project\train\train\o\combined.png\\"
]
TEMPLATE_NAMES = ['OH', 'OR', 'O', 'H', 'N', 'RO']

SINGLE_TEMPLATE_PATHS = [
r'C:\Users\rsurs\OneDrive\Documents\University_Study Material\3rd year\CV_LAB\Mini_Project\train\o',
r'C:\Users\rsurs\OneDrive\Documents\University_Study Material\3rd year\CV_LAB\Mini_Project\train\h',
r'C:\Users\rsurs\OneDrive\Documents\University_Study Material\3rd year\CV_LAB\Mini_Project\train\n',
]

STRUCTURES = [
    'struct1', 
    'struct4', 
    'struct5',
    'struct8',
    'struct13',
    'struct16',
    'struct19',
    'struct20',
    'struct22'
]

SINGLE_TEMPLATE_NAMES = ['O', 'H', 'N']

TRAIN_IMAGES = ['01.png', '09.png', '17.png', '25.png', '33.png']
PYRAMID_SIZES = range(20, 70, 10)
STEP = 5

def get_max_size(x_list, y_list):
    s = max([max(x_list), max(y_list)])
    return (s, s)

# Function to deskew an image
def deskew(img, SZ):
    m = cv2.moments(img)
    if abs(m['mu02']) < 1e-2:
        return img.copy()
    skew = m['mu11'] / m['mu02']
    M = np.float32([[1, skew, -0.5 * SZ * skew], [0, 1, 0]])
    img = cv2.warpAffine(img, M, (SZ, SZ))
    return img

# Function to crop and resize an image
def crop_and_resize_image(thresh_im, size_before_pad, pad):
    height, width = thresh_im.shape
    top_crop = 0
    bottom_crop = 0
    left_crop = 0
    right_crop = 0
    
    # Crop from the top
    for y in range(height):
        row = thresh_im[y, :]
        if np.count_nonzero(row) > 0:
            top_crop = y
            break
    
    # Crop from the bottom
    for y in reversed(range(height)):
        row = thresh_im[y, :]
        if np.count_nonzero(row) > 0:
            bottom_crop = y
            break
    
    # Crop from the left
    for x in range(width):
        col = thresh_im[:, x]
        if np.count_nonzero(col) > 0:
            left_crop = x
            break
    
    # Crop from the right
    for x in reversed(range(width)):
        col = thresh_im[:, x]
        if np.count_nonzero(col) > 0:
            right_crop = x
            break
    
    # Crop and resize the image
    fully_cropped = thresh_im[top_crop:bottom_crop, left_crop:right_crop]
    fully_cropped = cv2.resize(fully_cropped, (size_before_pad, size_before_pad))
    fully_cropped = cv2.copyMakeBorder(fully_cropped, pad, pad, pad, pad, cv2.BORDER_CONSTANT, 0)
    return fully_cropped

# Function to crop and create templates from images in a given path
def crop_and_make_templates(path):
    ims = []
    for i, image in enumerate(os.listdir(path)):
        if image.endswith('.png'):
            full_name = os.path.join(path, image)
            im = cv2.imread(full_name, 0)
            ret, im = cv2.threshold(im, 100, 255, cv2.THRESH_BINARY_INV)
            im = crop_and_resize_image(im, 40, 0)
            ims.append(im)
    return ims

# Function to stack templates and create a combined template image
def stack_templates(path, train_split=0.9):
    if not os.path.exists(path):
        print(f"Error: The directory '{path}' does not exist.")
        return

    ims = crop_and_make_templates(path)
    n_images = float(len(ims))
    
    # Apply Gaussian filter and normalize
    ims = [cv2.GaussianBlur(im, (5, 5), 5) / n_images for im in ims]
    
    final_image = ims[0]
    for im in ims[1:]:
        final_image = cv2.add(final_image, im)
    
    # Display the combined template image
    plt.imshow(final_image, cmap='gray')
    plt.show()


# Create and stack templates for 'train/r/', 'train/o/', 'train/h/', and 'train/n/'
stack_templates(r'C:\Users\rsurs\OneDrive\Documents\University_Study Material\3rd year\CV_LAB\Mini_Project\train\train\oh')
stack_templates(r'C:\Users\rsurs\OneDrive\Documents\University_Study Material\3rd year\CV_LAB\Mini_Project\train\train\h')
stack_templates(r'C:\Users\rsurs\OneDrive\Documents\University_Study Material\3rd year\CV_LAB\Mini_Project\train\train\n')
stack_templates(r'C:\Users\rsurs\OneDrive\Documents\University_Study Material\3rd year\CV_LAB\Mini_Project\train\train\oh')
stack_templates(r'C:\Users\rsurs\OneDrive\Documents\University_Study Material\3rd year\CV_LAB\Mini_Project\train\train\or')
stack_templates(r'C:\Users\rsurs\OneDrive\Documents\University_Study Material\3rd year\CV_LAB\Mini_Project\train\train\ro')
```

```python
#Training Model via LinearSVC,LogisticRegression,ANN-MLP

import os
import numpy as np
import cv2
from sklearn.svm import LinearSVC
from sklearn.linear_model import LogisticRegression
from sklearn.neural_network import MLPClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
import joblib

def extract_hog_features(image_path):
    image = cv2.imread(image_path, cv2.IMREAD_GRAYSCALE)
    # Resize image to ensure consistency
    image = cv2.resize(image, (128, 128))
    hog = cv2.HOGDescriptor()
    features = hog.compute(image).flatten()
    return features

def classify_bonds(bond_images, trained_classifier):
    """Classify the bonds using the trained classifier."""
    # Extract HOG features from bond_images
    bond_features = [extract_hog_features(bond_img) for bond_img in bond_images]

    # Predict bond types using the trained classifier
    bond_types = trained_classifier.predict(bond_features)

    return bond_types

# Paths and Lists
train_folder_path = (r'C:\Users\rsurs\OneDrive\Documents\University_Study Material\3rd year\CV_LAB\Mini_Project\train\train')
bond_types = os.listdir(train_folder_path)
hog_features_list = []
labels_list = []

# Extract Features and Labels
for bond_type in bond_types:
    bond_folder_path = os.path.join(train_folder_path, bond_type)
    for image_name in os.listdir(bond_folder_path):
        image_path = os.path.join(bond_folder_path, image_name)
        hog_features = extract_hog_features(image_path)
        hog_features_list.append(hog_features)
        labels_list.append(bond_type)

# Convert lists to arrays for consistency
hog_features_array = np.array(hog_features_list)
labels_array = np.array(labels_list)

# Train-Test Split
X_train, X_test, y_train, y_test = train_test_split(hog_features_array, labels_array, test_size=0.2, random_state=42)

# Train LinearSVC
clf = LinearSVC(max_iter=10000, random_state=42)
clf.fit(X_train, y_train)

# Evaluate
y_pred = clf.predict(X_test)
accuracy = accuracy_score(y_test, y_pred)
print(f"Test Accuracy for Linear SVC: {accuracy * 100:.2f}%")

# Save the trained classifier
joblib.dump(clf,r'C:\Users\rsurs\OneDrive\Documents\University_Study Material\3rd year\CV_LAB\Final_Project')


# Train Logistic Regression
# Note: 'multi_class' is set to 'ovr' for one-vs-rest approach which is commonly used for multi-class classification
clf = LogisticRegression(max_iter=10000, random_state=42, multi_class='ovr')
clf.fit(X_train, y_train)

# Evaluate
y_pred = clf.predict(X_test)
accuracy = accuracy_score(y_test, y_pred)
print(f"Test Accuracy for Logistic Regression: {accuracy * 100:.2f}%")

# Save the trained classifier
joblib.dump(clf, r'C:\Users\rsurs\OneDrive\Documents\University_Study Material\3rd year\CV_LAB\Final_Project')

nn_clf = MLPClassifier(hidden_layer_sizes=(128,), max_iter=10000, activation='relu', random_state=42)
nn_clf.fit(X_train, y_train)

# Evaluate the One-Layer Neural Network
y_pred = nn_clf.predict(X_test)
accuracy = accuracy_score(y_test, y_pred)
print(f"Test Accuracy for One-Layer Neural Network: {accuracy * 100:.2f}%")

# Save the trained One-Layer Neural Network classifier
joblib.dump(nn_clf, r'C:\Users\rsurs\OneDrive\Documents\University_Study Material\3rd year\CV_LAB\Final_Project')
```

```python
#Text Recognition

import cv2
import numpy as np
import matplotlib.pyplot as plt
from skimage.feature import hog
from sklearn.model_selection import train_test_split
from sklearn.metrics import f1_score
from sklearn.svm import SVC
from sklearn.linear_model import LogisticRegression
from sklearn.neural_network import MLPClassifier
import joblib

image = cv2.imread(r'C:\Users\rsurs\OneDrive\Documents\University_Study Material\3rd year\CV_LAB\Mini_Project\data\struct13\sd\struct13_01.png', cv2.IMREAD_COLOR)

templates = {
    "O": cv2.imread(r'C:\Users\rsurs\OneDrive\Documents\University_Study Material\3rd year\CV_LAB\Mini_Project\train\train\o\combined_all.png', 0),
    "H": cv2.imread(r'C:\Users\rsurs\OneDrive\Documents\University_Study Material\3rd year\CV_LAB\Mini_Project\train\train\h\combined_all.png', 0),
    "N": cv2.imread(r'C:\Users\rsurs\OneDrive\Documents\University_Study Material\3rd year\CV_LAB\Mini_Project\train\train\n\combined_all.png', 0),
    "OH": cv2.imread(r'C:\Users\rsurs\OneDrive\Documents\University_Study Material\3rd year\CV_LAB\Mini_Project\train\train\oh\combined_deskew.png', 0),
    "OR": cv2.imread(r'C:\Users\rsurs\OneDrive\Documents\University_Study Material\3rd year\CV_LAB\Mini_Project\train\train\or\combined_deskew.png', 0),
    "RO": cv2.imread(r'C:\Users\rsurs\OneDrive\Documents\University_Study Material\3rd year\CV_LAB\Mini_Project\train\train\ro\combined_deskew.png', 0),
}


for label, img in templates.items():
    if img is None:
        print(f"Template image for '{label}' could not be loaded. Check the file path.")
    else:
        print(f"Template image for '{label}' is loaded with shape: {img.shape}")
        
def apply_gaussian_blur(image, kernel_size):
    # Make sure the kernel size is odd
    if kernel_size % 2 == 0:
        kernel_size += 1
    return cv2.GaussianBlur(image, (kernel_size, kernel_size), 0)

stroke_width = 2  # example stroke width, should be measured
# kernel size equal to twice the stroke width, adjusted to be odd
kernel_size = stroke_width * 2 if (stroke_width * 2) % 2 == 1 else (stroke_width * 2) + 1

for label, template in templates.items():
    # Apply Gaussian blur to each template
    templates[label] = apply_gaussian_blur(template, kernel_size)
    
def template_matching(image, template):
    # Store matches, scores, and scales
    matches = []
    for scale in np.linspace(0.3, 3.0, 20):  # Scale from 30% to 300%
        resized_template = cv2.resize(template, None, fx=scale, fy=scale)
        result = cv2.matchTemplate(image, resized_template, cv2.TM_CCOEFF_NORMED)
        threshold = 0.5  # Tolerance level for matching
        locations = np.where(result >= threshold)
        for loc in zip(*locations[::-1]):  # Find locations above threshold
            matches.append((loc, result[loc], scale))

classifier = joblib.load(r'C:\Users\91636\Documents\Sem5\CV\JupyterCV\linearSVC_classifier.pkl')

test_image = cv2.imread(r'C:\Users\91636\Documents\Sem5\CV\JupyterCV\data\struct13\struct13_01.png', 0)
test_image = apply_gaussian_blur(test_image, kernel_size)

def non_max_suppression_fast(boxes, overlapThresh):
    # If there are no boxes, return an empty list
    if len(boxes) == 0:
        return []

    # Initialize the list of picked indexes
    pick = []

    # Grab the coordinates of the bounding boxes
    x1 = boxes[:,0]
    y1 = boxes[:,1]
    x2 = boxes[:,2]
    y2 = boxes[:,3]

    # Compute the area of the bounding boxes and sort the bounding
    # boxes by the bottom-right y-coordinate of the bounding box
    area = (x2 - x1 + 1) * (y2 - y1 + 1)
    idxs = np.argsort(y2)

    # Keep looping while some indexes still remain in the indexes list
    while len(idxs) > 0:
        # Grab the last index in the indexes list and add the
        # index value to the list of picked indexes
        last = len(idxs) - 1
        i = idxs[last]
        pick.append(i)

        # Find the largest (x, y) coordinates for the start of
        # the bounding box and the smallest (x, y) coordinates
        # for the end of the bounding box
        xx1 = np.maximum(x1[i], x1[idxs[:last]])
        yy1 = np.maximum(y1[i], y1[idxs[:last]])
        xx2 = np.minimum(x2[i], x2[idxs[:last]])
        yy2 = np.minimum(y2[i], y2[idxs[:last]])

        # Compute the width and height of the bounding box
        w = np.maximum(0, xx2 - xx1 + 1)
        h = np.maximum(0, yy2 - yy1 + 1)

        # Compute the ratio of overlap
        overlap = (w * h) / area[idxs[:last]]

        # Delete all indexes from the index list that have
        idxs = np.delete(idxs, np.concatenate(([last],
            np.where(overlap > overlapThresh)[0])))

    # Return only the bounding boxes that were picked
    return boxes[pick]


# Define the scale range and number of scales
start_scale = 0.3  # Starting scale (30% of original size)
end_scale = 3.0    # Ending scale (300% of original size)
num_scales = 20    # Number of scales to consider

def detect_text(image, templates, classifier):
    detections = []
    for label, template in templates.items():
        for scale in np.linspace(start_scale, end_scale, num_scales):  # Adjust as needed
            resized_template = cv2.resize(template, None, fx=scale, fy=scale)
            result = cv2.matchTemplate(image, resized_template, cv2.TM_CCOEFF_NORMED)
            threshold = 0.77  # Tolerance level for matching
            locations = np.where(result >= threshold)
            if locations[0].size == 0:
                print(f"No matches found for template '{label}' at scale {scale:.2f}")
                continue  # Skip to the next scale if no matches found
            for loc in zip(*locations[::-1]):  # Find locations above threshold
                # Extract the region from the original image
                start_x = int(loc[0] / scale)
                start_y = int(loc[1] / scale)
                end_x = int((loc[0] + template.shape[1]) / scale)
                end_y = int((loc[1] + template.shape[0]) / scale)
                if start_x >= image.shape[1] or start_y >= image.shape[0] or end_x <= 0 or end_y <= 0:
                    print(f"Invalid region for template '{label}' at location {loc} and scale {scale:.2f}")
                    continue  # Skip to the next match if the region is invalid
                region = image[start_y:end_y, start_x:end_x]
                if region.size == 0:
                    print(f"Empty region extracted for template '{label}' at location {loc} and scale {scale:.2f}")
                    continue  # Skip to the next match if the extracted region is empty
                
                # Extract features and predict using the classifier
                features = extract_hog_features(region)
                prediction = classifier.predict([features])[0]
                
                if prediction == label:  # Check if prediction matches the template label
                    detections.append((start_x, start_y, end_x, end_y))

    # Apply non-maximum suppression to the detections
    detections = np.array(detections)
    nms_detections = non_max_suppression_fast(detections, 0.3)  # Example overlap threshold
    return nms_detections


# Detect text in the test image
detected_regions = detect_text(test_image, templates, classifier)

# Draw the detected regions on the image
for (startX, startY, endX, endY) in detected_regions:
    cv2.rectangle(test_image, (startX, startY), (endX, endY), (255, 0, 0), 2)
```

```python
#Intersect Function

def ccw(A,B,C):
	return (C[0]-A[0])*(B[1]-A[1]) > (B[0]-A[0])*(C[1]-A[1])

def intersects(A,B,C,D):
	return ccw(A,C,D) != ccw(B,C,D) and ccw(A,B,C) != ccw(A,B,D)
```

```python
import cv2
import keras_ocr
import numpy as np
import math
import matplotlib.pyplot as plt

# Initialize the keras-ocr pipeline
pipeline = keras_ocr.pipeline.Pipeline()

def midpoint(pt1, pt2):
    return (int((pt1[0] + pt2[0]) / 2), int((pt1[1] + pt2[1]) / 2))

def remove_text_from_image(image_path, pipeline):
    # Read the image from the path
    img = keras_ocr.tools.read(image_path)
    
    # Generate (word, box) tuples
    prediction_groups = pipeline.recognize([img])
    
    # Create a mask for the detected text
    mask = np.zeros(img.shape[:2], dtype="uint8")
    for word, box in prediction_groups[0]:
        # Calculate the midpoint of the bottom line of the text box
        start_point = midpoint(box[1], box[2])
        end_point = midpoint(box[0], box[3])
        
        # Estimate thickness based on the height of the box
        thickness = int(math.sqrt((box[2][0] - box[1][0]) ** 2 + (box[2][1] - box[1][1]) ** 2))
        
        # Draw the line on the mask
        cv2.line(mask, start_point, end_point, 255, thickness)
    
    # Inpaint the text on the original image using the mask
    inpainted_image = cv2.inpaint(img, mask, 7, cv2.INPAINT_NS)
    
    # Return the inpainted image
    return inpainted_image

# Example usage:
image_path = r'C:\Users\rsurs\OneDrive\Documents\University_Study Material\3rd year\CV_LAB\Mini_Project\data\struct13\sd\struct13_01.png'  # Update with the path to your image
inpainted_image = remove_text_from_image(image_path, pipeline)

# Convert to RGB for display
inpainted_image_rgb = cv2.cvtColor(inpainted_image, cv2.COLOR_BGR2RGB)
plt.imshow(inpainted_image_rgb)
plt.axis('off')  # Hide the axis
plt.show()

```
